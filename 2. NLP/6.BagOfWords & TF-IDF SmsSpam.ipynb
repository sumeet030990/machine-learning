# Implementation of Bag of Words & TF-IDF: SMS Spam/Ham identifier

#### Step 1. Data ingestion

```python
import pandas as pd
import nltk

df = pd.read_csv("resources/6.sms_spam.csv", encoding='latin-1')
df.drop(["Unnamed: 2",	"Unnamed: 3",	"Unnamed: 4"], axis=1, inplace=True)
df.rename({"v1":"label", "v2": "message"}, axis=1, inplace=True)
df
```

#### Step 2. Data Cleaning and Preprocessing  

```python
import re
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer


stemmer = SnowballStemmer('english')

nltk.download('stopwords')
```

```python
corpus = []

for message in df.message:
  reviews = re.sub('[^a-zA-Z]',' ', message)
  reviews = reviews.lower()
  reviews = reviews.split() ## converting sentence to words
  reviews = [stemmer.stem(review) for review in reviews if review not in set(stopwords.words('english'))]
  reviews = (' ').join(reviews)
  corpus.append(reviews)
```

```python
corpus
```

## Bag of words

```python
## Create bag of words model

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, binary=True)
```

```python
X = cv.fit_transform(corpus).toarray()
X.shape
```

```python
cv.vocabulary_ ## words with their indexes
```

### Usign N-Gram in CountVectorizer

#### unigram

```python
cv = CountVectorizer(max_features=2500, binary=True, ngram_range=(1,1))
X = cv.fit_transform(corpus).toarray()
cv.vocabulary_

## the output ll be the same as the vocabullary ll contain single word onky
```

#### Bigram: Vocabulary ll contain single words and combination of 2 words also

```python
cv = CountVectorizer(max_features=2500, binary=True, ngram_range=(1,2))
X = cv.fit_transform(corpus).toarray()
cv.vocabulary_

## the output ll contain single and double words eg 'free entri'
```

#### Bigram: Vocabulary ll contain single words and combination of 3 words

```python
cv = CountVectorizer(max_features=2500, binary=True, ngram_range=(1,3))
X = cv.fit_transform(corpus).toarray()
cv.vocabulary_

## the output ll contain single words and combination of three words 'free entri wkli'
```

Combination of Bigram, Trigram

```python
cv = CountVectorizer(max_features=2500, binary=True, ngram_range=(2,3))
X = cv.fit_transform(corpus).toarray()
cv.vocabulary_

## the output ll contain bigram and trigram: 'per request', 'oru minnaminungint nurungu'
```

##### There can be any combination of grams 
- 1,1 
- 2,2 
- 3,3 
- 1,2 
- 1,3 
- 2,3

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_idf_vectorizer = TfidfVectorizer(max_features=100)
```

```python
X = tf_idf_vectorizer.fit_transform(corpus).toarray()

tf_idf_vectorizer.vocabulary_
```
